```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import  linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import make_pipeline
from sklearn.utils import resample
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from sklearn.model_selection import GridSearchCV
import math as ma
import seaborn as sns
import autograd.numpy as np
from sklearn.decomposition import PCA


#Data set
from sklearn.datasets import load_wine
from sklearn.neural_network import MLPClassifier
```

```python
A = pd.read_csv("Wine_PCA_7_features")
W = pd.read_csv("Wine_PCA_5_features")
C = pd.read_csv("Wine_PCA_2_features")
```

```python
wine = load_wine()
```

```python
np.random.seed(645)

z = wine.target.copy()
t = wine.target.copy()
e = wine.target.copy()
y = wine.target.copy()
X = wine.data.copy()
```

```python
np.random.seed(645)

#Data without PCA
X_train, X_test, z_train, z_test = train_test_split(X,z, test_size=0.2)

#80% Variance through 5 PCA features
W_train, W_test, t_train, t_test = train_test_split(W,t, test_size=0.2)

#90% Variance through 7 PCA Features
A_train, A_test, y_train, y_test = train_test_split(A,y, test_size=0.2)


#2 Pricipal components to check
C_train, C_test, e_train, e_test = train_test_split(C,e, test_size=0.2)


```

# Scaling Data

```python
scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)


scaler_W = StandardScaler()
W_train = scaler_W.fit_transform(W_train)
W_test = scaler_W.transform(W_test)


scaler_A = StandardScaler()
A_train = scaler_A.fit_transform(A_train)
A_test = scaler_A.transform(A_test)


scaler_C = StandardScaler()
C_train = scaler_C.fit_transform(C_train)
C_test = scaler_C.transform(C_test)

```

```python
np.random.seed(645)
acc_no_pca = []
acc_5_feat = []
acc_7_feat = []
acc_2_feat = []


acc_no_pca_train = []
acc_5_feat_train = []
acc_7_feat_train = []
acc_2_feat_train = []




for i in range(1,251):
    
    
    clf_no_pca = MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(X_train, z_train)
    clf_no_pca.predict(X_test)

    clf_5_features = MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(W_train, t_train)
    clf_5_features.predict(W_test)

    clf_7_features =  MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(A_train, y_train)
    clf_7_features.predict(A_test)
    
    clf_2_features =  MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(C_train, e_train)
    clf_2_features.predict(C_test)
    
    acc_no_pca.append(clf_no_pca.score(X_test,z_test))
    acc_5_feat.append(clf_5_features.score(W_test,t_test))
    acc_7_feat.append(clf_7_features.score(A_test,y_test))
    acc_2_feat.append(clf_2_features.score(C_test,e_test))
    
    acc_no_pca_train.append(clf_no_pca.score(X_train,z_train))
    acc_5_feat_train.append(clf_5_features.score(W_train,t_train))
    acc_7_feat_train.append(clf_7_features.score(A_train,y_train))
    acc_2_feat_train.append(clf_2_features.score(C_train,e_train))
    
```

```python
plt.plot(np.arange(1,251), acc_no_pca, label='No PCA 13 features')
plt.plot(np.arange(1,251), acc_5_feat, label='5 PC')
plt.plot(np.arange(1,251), acc_7_feat, label='7 PC')
plt.plot(np.arange(1,251), acc_2_feat, label='2 PC')


plt.legend()
plt.xlabel("n_epochs")
plt.ylabel("accuracy")
plt.ylim(0.8,1.01)
plt.title("accuracy score PCA vs Non PCA")
plt.show()
```
