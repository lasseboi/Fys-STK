```python


import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn import  linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import make_pipeline
from sklearn.utils import resample
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from sklearn.model_selection import GridSearchCV
import math as ma
import seaborn as sns

from sklearn.decomposition import PCA


#Data set
from sklearn.datasets import load_breast_cancer
from sklearn.neural_network import MLPClassifier

```

```python
X = pd.read_csv("WBCD_PCA_all_features")


BC= load_breast_cancer()

z = BC.target.copy()
a = BC.target.copy()
b = BC.target.copy()
c = BC.target.copy()
d = BC.target.copy()
e = BC.target.copy()
f = BC.target.copy()
```

```python
PC_labels = X.columns



```

```python
#5PC
A = X[PC_labels[0:5]]
#6PC
B = X[PC_labels[0:6]]
#7 PC
C = X[PC_labels[0:7]]
# 8PC
D = X[PC_labels[0:8]]
#9 PC
E = X[PC_labels[0:9]]
# 10 PC
F = X[PC_labels[0:10]]
```

```python
np.random.seed(645)

#all PCA features
X_train, X_test, z_train, z_test = train_test_split(X,z, test_size=0.2)

# 5 PCA features
A_train, A_test, a_train, a_test = train_test_split(A,a, test_size=0.2)

#6 PCA features
B_train, B_test, b_train, b_test = train_test_split(B,b, test_size=0.2)

# 7 PCA Features
C_train, C_test, c_train, c_test = train_test_split(C,c, test_size=0.2)

#8 PCA features
D_train, D_test, d_train, d_test = train_test_split(D,d, test_size=0.2)

#9 PCA features
E_train, E_test, e_train, e_test = train_test_split(E,e, test_size=0.2)


#10 Pricipal components to check
F_train, F_test, f_train, f_test = train_test_split(F,f, test_size=0.2)
```

```python



np.random.seed(645)



scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

scaler_A = StandardScaler()
A_train = scaler_A.fit_transform(A_train)
A_test = scaler_A.transform(A_test)



scaler_B = StandardScaler()
B_train = scaler_B.fit_transform(B_train)
B_test = scaler_B.transform(B_test)




scaler_C = StandardScaler()
C_train = scaler_C.fit_transform(C_train)
C_test = scaler_C.transform(C_test)

scaler_D = StandardScaler()
D_train = scaler_D.fit_transform(D_train)
D_test = scaler_D.transform(D_test)


scaler_E = StandardScaler()
E_train = scaler_E.fit_transform(E_train)
E_test = scaler_E.transform(E_test)

scaler_F = StandardScaler()
F_train = scaler_F.fit_transform(F_train)
F_test = scaler_F.transform(F_test)

```

```python
np.random.seed(645)
acc_no_pca = []
acc_5_feat = []
acc_6_feat = []
acc_7_feat = []
acc_8_feat = []
acc_9_feat = []
acc_10_feat = []


acc_no_pca_train = []
acc_5_feat_train = []
acc_6_feat_train = []
acc_7_feat_train = []
acc_8_feat_train = []
acc_9_feat_train = []
acc_10_feat_train = []




for i in range(1,251):
    
    
    clf_no_pca = MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(X_train, z_train)
    clf_no_pca.predict(X_test)

    clf_5_features = MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(A_train, a_train)
    clf_5_features.predict(A_test)

    clf_6_features =  MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(B_train, b_train)
    clf_6_features.predict(B_test)
    
    clf_7_features =  MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(C_train, c_train)
    clf_7_features.predict(C_test)
    
    clf_8_features =  MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(D_train, d_train)
    clf_8_features.predict(D_test)
    
    
    clf_9_features =  MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(E_train, e_train)
    clf_9_features.predict(E_test)
    
    clf_10_features =  MLPClassifier(random_state=1, max_iter=i,solver="adam").fit(F_train, f_train)
    clf_10_features.predict(F_test)
    
    acc_no_pca.append(clf_no_pca.score(X_test,z_test))
    acc_5_feat.append(clf_5_features.score(A_test,a_test))
    acc_6_feat.append(clf_6_features.score(B_test,b_test))
    acc_7_feat.append(clf_7_features.score(C_test,c_test))
    acc_8_feat.append(clf_8_features.score(D_test,d_test))
    acc_9_feat.append(clf_9_features.score(E_test,e_test))
    acc_10_feat.append(clf_10_features.score(F_test,f_test))
    
    acc_no_pca_train.append(clf_no_pca.score(X_train,z_train))
    acc_5_feat_train.append(clf_5_features.score(A_train,a_train))
    acc_6_feat_train.append(clf_6_features.score(B_train,b_train))
    acc_7_feat_train.append(clf_7_features.score(C_train,c_train))
    acc_8_feat_train.append(clf_8_features.score(D_train,d_train))
    acc_9_feat_train.append(clf_9_features.score(E_train,e_train))
    acc_10_feat_train.append(clf_10_features.score(F_train,f_train))
    
```

```python
plt.plot(np.arange(1,251), acc_no_pca, label=' 13 PC features')
plt.plot(np.arange(1,251), acc_5_feat, label='5 PC')
plt.plot(np.arange(1,251), acc_6_feat, label='6 PC')
plt.plot(np.arange(1,251), acc_7_feat, label='7 PC')
plt.plot(np.arange(1,251), acc_8_feat, label='8 PC')
plt.plot(np.arange(1,251), acc_9_feat, label='9 PC')
plt.plot(np.arange(1,251), acc_10_feat, label='10 PC')


plt.legend()
plt.xlabel("n_epochs")
plt.ylabel("accuracy")
plt.ylim(0.8,1.01)
plt.title("accuracy score based on Principal components")
plt.show()
```
